```python
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
#Read words file
file = open('D:\\Facundo\\Code\\addresscompare\\data\\Properties_name_stats.csv', mode='r')
#file = open('C:\\code\\addresscompare\\data\\Properties_name_stats.csv', mode='r')

words_reader = csv.reader(file)

headers_words = next(words_reader)
print(headers_words)

mydict_idf = {rows[1]:rows[4] for rows in words_reader}
```

```python
#Read hotels file
file = open('D:\\Facundo\\Code\\addresscompare\\data\\propertylong.csv', mode='r')
#file = open('C:\\code\\addresscompare\\data\\propertylong.csv', mode='r')

hotels_reader = csv.reader(file,delimiter='\t',quoting=csv.QUOTE_MINIMAL)

headers_hotels = next(hotels_reader)

rows_hotels = []
for row in hotels_reader:
    rows_hotels.append(row)

#print(rows_hotels)
```

```python
def jacIdf_coefficient(a, b):
    
    if not len(a) or not len(b): return 0.0
    
    #quick case for true duplicates
    if a == b: return 1.0

    #if a != b, and a or b are single chars, then they can't possibly match
    if len(a) == 1 or len(b) == 1: return 0.0
    
    #use python list comprehension, preferred over list.append() 
    a_bigram_list = a.split()
    b_bigram_list = b.split()
    
    a_bigram_list.sort()
    b_bigram_list.sort()
    
    #print(a)
    #print(b)
    # assignments to save function calls
    lena = len(a_bigram_list)
    lenb = len(b_bigram_list)
    # initialize match counters
    i = j = 0
    matches = union = 0.0
    
    #Calculate union
    for a_word in a_bigram_list:
        if a_word in mydict_idf.keys():
            union += float(mydict_idf[a_word])
        else:
            #One appearence case
            union += 14.563995941228978

    for b_word in b_bigram_list:
        if b_word in mydict_idf.keys():
            union += float(mydict_idf[b_word])
        else:
            #One appearence case
            union += 14.563995941228978
    
    
    #Calculate intersection
    while (i < lena and j < lenb):
        if a_bigram_list[i] == b_bigram_list[j]:
            if a_bigram_list[i] in mydict_idf.keys():
                matches += float(mydict_idf[a_bigram_list[i]])
            else:
                #One appearence case
                union += 14.563995941228978
            
            i += 1
            j += 1
        elif a_bigram_list[i] < b_bigram_list[j]:
            i += 1
        else:
            j += 1
    
    score = float(matches)/float(union - matches)
    return score
```
