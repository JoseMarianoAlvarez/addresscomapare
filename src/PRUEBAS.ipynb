```python
def jacIdf_coefficient(a, b):
    
    if not len(a) or not len(b): return 0.0
    """ quick case for true duplicates """
    
    if a == b: return 1.0
    """ if a != b, and a or b are single chars, then they can't possibly match """
    
    if len(a) == 1 or len(b) == 1: return 0.0
    
    """ use python list comprehension, preferred over list.append() """
    a_bigram_list = a.split()
    b_bigram_list = b.split()
    
    a_bigram_list.sort()
    b_bigram_list.sort()
    
    #print(a)
    #print(b)
    # assignments to save function calls
    lena = len(a_bigram_list)
    lenb = len(b_bigram_list)
    # initialize match counters
    i = j = 0
    matches = union = 0.0
    
    #Calculate union
    for a_word in a_bigram_list:
        union += float(get_idf(a_word))
    for b_word in b_bigram_list:
        union += float(get_idf(b_word))
    
    #Calculate intersection
    while (i < lena and j < lenb):
        if a_bigram_list[i] == b_bigram_list[j]:
            matches += float(get_idf(a_bigram_list[i]))
            i += 1
            j += 1
        elif a_bigram_list[i] < b_bigram_list[j]:
            i += 1
        else:
            j += 1
    
    score = float(matches)/float(union - matches)
    return score
```

```python
def jacFQ_coefficient(a, b):
    
    if not len(a) or not len(b): return 0.0
    """ quick case for true duplicates """
    
    if a == b: return 1.0
    """ if a != b, and a or b are single chars, then they can't possibly match """
    
    if len(a) == 1 or len(b) == 1: return 0.0
    
    """ use python list comprehension, preferred over list.append() """
    a_bigram_list = a.split()
    b_bigram_list = b.split()
    
    a_bigram_list.sort()
    b_bigram_list.sort()
    
    #print(a)
    #print(b)
    # assignments to save function calls
    lena = len(a_bigram_list)
    lenb = len(b_bigram_list)
    # initialize match counters
    i = j = 0
    matches = union = 0.0
    
    #Calculate union
    for a_word in a_bigram_list:
        union += float(get_FQ(a_word))
    for b_word in b_bigram_list:
        union += float(get_FQ(b_word))

    #Calculate intersection
    while (i < lena and j < lenb):
        if a_bigram_list[i] == b_bigram_list[j]:
            matches += float(get_FQ(a_bigram_list[i]))
            i += 1
            j += 1
        elif a_bigram_list[i] < b_bigram_list[j]:
            i += 1
        else:
            j += 1
    
    score = float(matches)/float(union - matches)
    return score
```

```python
def get_idf(st):
    for i in range(len(rows_words)):
        if (rows_words[i][1] == st):
            return rows_words[i][4]
    
    # Default return idf 1 appearance.
    return rows_words[len(rows_words)-1][4]
```

```python
def get_FQ(st):
    for i in range(len(rows_words)):
        if (rows_words[i][1] == st):
            return rows_words[i][3]
    
    # Default return idf 1 appearance.
    return rows_words[len(rows_words)-1][3]
```
